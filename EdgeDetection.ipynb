```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
%matplotlib inline
```

```
image = cv2.imread('taj.jpg')

## copy
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```
## gray scale image
image_gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

plt.imshow(image_gray, cmap='gray')
```

```
## to detect all the vertical lines
sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])

filter_image = cv2.filter2D(image_gray, -1, sobel_x)

plt.imshow(filter_image, cmap='gray')
```

```
retva, binary_image = cv2.threshold(filter_image, 100, 255, cv2.THRESH_BINARY)

plt.imshow(binary_image, cmap='gray')
```

```
plt.imshow(image_gray, cmap="gray")
```

```
lower = 120
upper = 240

canny_image = cv2.Canny(image_gray, lower, upper)

plt.imshow(canny_image, cmap="gray")
```
