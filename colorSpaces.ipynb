```
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```
image = cv2.imread('looney.jpg')

print(image.shape)

img_copy = np.copy(image)

img_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(img_copy)
```

```
# RGB Channel
r = img_copy[:,:,0] # take all the values in length and breadth and take the 0th value from 3rd colum (the red value)
g = img_copy[:,:,1]
b = img_copy[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title('red value')
ax1.imshow(r, cmap='gray')

ax2.set_title('green value')
ax2.imshow(g, cmap='gray')

ax3.set_title('blue value')
ax3.imshow(b, cmap='gray')
```

So, not much of a difference. So, we'll not be able to effectively select Mr. Bugs Bunny !

```
hsv = cv2.cvtColor(img_copy, cv2.COLOR_RGB2HSV)

# HSV Channel
h = hsv[:,:,0] # take all the values in length and breadth and take the 0th value from 3rd colum (the hue value)
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1, ax2, ax3) = plt.subplots(1,3,figsize=(20,10))

ax1.set_title('hue value')
ax1.imshow(h, cmap='gray')

ax2.set_title('saturation value')
ax2.imshow(s, cmap='gray')

ax3.set_title('value')
ax3.imshow(v, cmap='gray')
```

```
lower_hue = np.array([100, 0, 0])
upper_hue = np.array([180, 255,255])
```

```
mask = cv2.inRange(hsv, lower_hue, upper_hue)

masked_image = np.copy(img_copy)
masked_image[mask == 0] = [0,0,0]

plt.imshow(masked_image)
```
